```python
from ta import add_all_ta_features
from ta.utils import dropna
from ta.volatility import BollingerBands
from ta.trend import ADXIndicator
from ta.volatility import AverageTrueRange
from ta.trend import SMAIndicator
from ta.momentum import RSIIndicator
from ta.volume import VolumeWeightedAveragePrice
```

```python
import yfinance as yf

import math
import pandas as pd
import numpy as np
```

```python
stock_symbol = "AAPL"
start = "2022-01-01"
end = "2022-09-11"
period = "1y"
```

```python
stock_data = yf.Ticker(stock_symbol)
hist_stock = stock_data.history(start = start, end = end)
hist_stock["Close"].plot()
```

```python
import matplotlib.pyplot as plt
```

```python
# RSI Indicator - assuming window is 14
D = RSIIndicator(close = hist_stock['Close'], window = 14, fillna = False)
RSI = D.rsi()
round(RSI, 1).plot()
```

```python
# SMA - assuming window is 14
C = SMAIndicator(close = hist_stock['Close'], window = 14, fillna = False)
MA = C.sma_indicator()
round(MA, 1).plot()
```

```python
# ATR - assuming window is 14
B = AverageTrueRange(high = hist_stock['High'], low= hist_stock['Low'], close= hist_stock['Close'], window= 14, fillna= False)
ATR = B.average_true_range()
round(ATR,1).plot()
```
